```python
import pandas as pd
# get the values for each column
filename1 = r"./kafka_2_1(30000).csv"
filename2 = r"./kafka_2_2(60000).csv"
filename3 = r"./kafka_2_3(100000).csv"

def csv_pd(fn):
    data_kafka = pd.read_csv(fn,header=None)
    data = pd.read_csv(r"./iot_static_data.csv")
    data_kafka.columns=data.columns

    data_kafka_label = data_kafka['Class']  # Class
    data_kafka_all = data_kafka.iloc[:,:-1] # include all num and Source
    data_kafka_num = data_kafka_all.iloc[:,:-1]  # all num
    data_kafka_source = data_kafka_all['Source'] # Source

    # extract num features values from kafka data
    for j in range(len(data_kafka_num.columns)):
        for i in range(len(data_kafka_num.index)):
            # get values from strings
            str_item = str(data_kafka_num.iloc[i][j])
            ii = str_item.index(':')
            num_item = float(str_item[ii+1:].strip())
            data_kafka_num.iloc[i][j] = num_item

    # extract source and Class feature values from kafaka data
    for i in range(len(data_kafka_source.index)):
    # get values from strings
        str_item = str(data_kafka_source[i])
        ii = str_item.index(':')
        str_item = str(str_item[ii+1:].strip())
        data_kafka_source.loc[i,0] = str_item

    for i in range(len(data_kafka_label.index)):
    # get values from strings
        str_item = str(data_kafka_label[i])
        ii = str_item.index(':')
        str_item = str(str_item[ii+1:].strip().strip('}'))
        data_kafka_label.loc[i,0] = str_item

    return data_kafka_label, data_kafka_num, data_kafka_source
```

```python
data_kafka_label_1, data_kafka_num_1, data_kafka_source_1 = csv_pd(filename1)
```

```python
data_kafka_label_2, data_kafka_num_2, data_kafka_source_2 = csv_pd(filename2)
```

```python
data_kafka_label_3, data_kafka_num_3, data_kafka_source_3 = csv_pd(filename3)
```

```python
# -------------- concat data by row -----------------------
result_kafka_label = pd.concat([data_kafka_label_1,data_kafka_label_2,data_kafka_label_3])
result_kafka = pd.concat([data_kafka_num_1,data_kafka_num_2,data_kafka_num_3])
result_kafka_source = pd.concat([data_kafka_source_1,data_kafka_source_2,data_kafka_source_3])

result_kafka_label = result_kafka_label.reset_index(drop=True)
result_kafka = result_kafka.reset_index(drop=True)
result_kafka_source = result_kafka_source.reset_index(drop=True)
```
